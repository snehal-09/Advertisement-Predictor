## lower limit =Q1 - 1.5 * IQR
## Upper limit=Q3 + 1.5 * IQR 
## If Value not lie between this value it is outlier

```python
# Find Q1 and Q3 
Q1 = df['Newspaper'].quantile(0.25)
print('Lower quantile is',Q1)

Q3 = df['Newspaper'].quantile(0.75)
print('Upper quantile is',Q3)
```

```python
# Find IQR 
IQR = Q3-Q1 
print("IQR is",IQR)
```

```python
# Find Upper And Lower limit 
Lower_limit=Q1-1.5*IQR 
print("Lower limit is",Lower_limit)

Upper_limit=Q3+1.5*IQR 
print("Upper limit is",Upper_limit)
```

```python
# find Value above upper limit 
df.loc[df['Newspaper']>Upper_limit]
```

```python
# find Value below lower limit 
df.loc[df['Newspaper']<Lower_limit]
```

```python
len(df.loc[df['Newspaper']>Upper_limit])/len(df)
```

```python
# since data of Newspaper is not normal so we will replace outlier with medium

df.loc[df['Newspaper']>Upper_limit,'Newspaper']=df['Newspaper'].median()
```

```python
# now check Outliers

sns.boxplot(x=df.Newspaper)
```

```python
sns.boxplot(x=df["Newspaper"])
plt.show()
```

# Model Building

### 1.Split data into X and Y

```python
x=df.drop('Sales',axis=1)   # means include onle TV(x1) ,Radio(x2) , Newspaper(x3) = Input Variable
y=df['Sales']    # Output variable
```

```python
x
```

```python
y
```

### 2.Split data for training and testing

* sklearn: scientific library which include all package related to ML,data cleaning and feature engineering

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=4)
# test_size=0.30 = 30% data for testing
#random_state=4 makes sure that every time you run the code, the data is split the same way into training and testing sets.
#If you don’t set random_state, the split will be different each time, which can change your results.
```

```python
x_train.shape  
```

* out of 200 data 140 is for train means 70% and remaning 30% data for testing

```python
x_test.shape
```

* 60 =num of rows , 3=num of columns

```python
y_test.shape
```

```python
y_train.shape
```

### 3.Actual model building

```python
# 1.Import linear Regression model 
from sklearn.linear_model import LinearRegression

# 2.Initialize model 
model=LinearRegression()

# 3.Train model with x train and y_train
model.fit(x_train,y_train)      # fit =Train model using training data
```

```python
# 4.Make prediction/test model 
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(y_test, y_pred))  # low MAe better performance
print("MSE:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))    # good whe R2 close to 1
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Root Mean Squared Error:", rmse)   # low RMSE better prediction
```

* for check model done correct prediction we compare y_pred with y_test (y_test are actual value of x_test)

```python
# 5.Slope/coefficient of x 
model.coef_
```

* m1= 0.04539939  ,m2=0.17757031   ,m3=0.00502354

```python
# 6.intercept (c)
model.intercept_
```

* c=3.090172035290914

### Final result 
* y=m1x1 + m2x2 + m3x3 +....+c
* y=0.04539939* x1(TV) + 0.17757031* x2(Radio) + 0.00502354* x3(Newspaper) + 3.090172035290914
* This is the best fit line
